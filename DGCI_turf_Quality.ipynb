```
import numpy as np 
import pandas as pd
```

```
### Importing the files name from the training CSV file. 
### Preparing a Numpy array with the file extension ".jpg" so that a loop can be 
###used to extract the images 

import pandas as pd
name_df = pd.read_csv('train.csv')
name = name_df.Id 
DGCI = name_df.DGCI

file = list()
for i in range(len(name)):
  l = str(name[i]) + str('.jpg')
  file.append(l)
```

```
## View the image through PIL(Python Image Library) and looking for the discrepancy
## in the shape and size of images. 

from PIL import Image

std_size = (400,300)
count = 0

for i in range(len(file)):
  im = Image.open(file[i])
  sz = im.size 
  if sz != std_size :
    print(file[i])
    count = count + 1   
print("Total number of files with non standard dimensions are: " + str(count))

```

```
## If there is no non-standard dimensions file, a loop is started to convert the 
## image into a numpy array. Green index value for each of the pixel is collected 
## to process further. Total number of images in 792.


### A nested list which contains the Green pixel values for the of the images.
### The length of the images is 792, and each element is the 300x400x3 matrix.
### R and B Column of the matrix is Zero and only Green values were picked from
### each images


images = list()   
from PIL import Image
for i in range(len(file)):
  green_mat = np.zeros((300,400,1))
  
  img = Image.open(file[i])
  array = np.array(img)
  for x in range(300):
    for y in range(400):
      green_mat[x][y] = array[x][y][1]
  images.append(green_mat)
  
images = np.array(images)
```

```
### Normalizing the dataset.


for i in range(len(images)):
  px = list()
  
  mtx = images[i]
  mean = mtx.mean()
  std = mtx.std()
  images[i]= (mtx-mean)/std
  
```

```
y_train = np.array(DGCI)
type(y_train)
```

```
shape= [300,400,1]
shape = tuple(shape)
print(shape)
```

```
### Creating a Neural Network


import numpy as np 
import keras 
from keras.datasets import mnist 
from keras.models import Model 
from keras.layers import Dense, Input
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten 
from keras import backend as k 
from keras import Sequential
from keras import optimizers
from keras.layers import Dense, Dropout, Flatten, Activation, BatchNormalization, regularizers
from keras.layers.noise import GaussianNoise

```

```
model = Sequential()
model.add(Conv2D(8, kernel_size=(5, 5), strides=(1, 1),activation='relu',input_shape=shape))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Dropout(0.4))
model.add(BatchNormalization())
model.add(Conv2D(16, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Dropout(0.4))
model.add(BatchNormalization())
# model.add(Conv2D(32, (5, 5), activation='relu'))
# model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```
model.compile(loss='mse',optimizer='adam', metrics=['mse'])
model.fit(images, y_train,batch_size=100,epochs=10, validation_split= 0.12)
```
